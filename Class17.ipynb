#### Exploratory Data Analysis EDA

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('Pokemon.csv', index_col = 0, encoding = 'ISO-8859-1')
df1.head(10)
```

```python
df1.describe()
```

```python
df1.shape
```

```python
df1.size
```

```python
g = plt.hist(df1['speed'],histtype = 'bar', ec = 'black',bins = 15) # bins give the number of partitions and ec give the outline col of hist
g = plt.xlabel('Pokemon Speed')
g = plt.ylabel('Frequency')
plt.show()
```

##### Seaborn is built on matplotlib but provides high level interface

```python
# distplot,barplot,violinplot,jointplot,regplot,pairplot,boxplot,swarmplot,factorplot
```

```python
sns.set_style
sns.distplot(df1['speed'])
```

```python
sns.lmplot(x='attack', y='defense', data = df1)
```

```python
sns.lmplot(x='attack', y='defense', data = df1, fit_reg = False)
```

```python
sns.lmplot(x='attack', y='defense', data = df1, fit_reg = False, hue = 'legendary')
```

```python
df1.columns
```

```python
g = sns.catplot(x = 'type1',
                   y = 'attack',
                   data = df1,
                   hue = 'legendary',
                   col = 'legendary',
                   kind = 'swarm')
g.set_xticklabels(rotation = -45)
```

```python
sns.boxplot(data = df1)
```

```python
sns.jointplot(x = 'attack', y='defense',data=df1)
```
