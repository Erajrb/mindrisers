```python
t = np.arange(0.,5.,0.2)
#red dashes,blue squares and greent triangles
plt.figure(figsize=(4,3))
plt.plot(t,t,'r--',t,t**2,'bs',t,t*3,'g^')   #plt.plot(x,x,'red dashes',x,x**2,'blue square',x,x*3,'green triangle')
```

```python
data = {
        'a' : np.arange(50),
        'c' : np.random.randint(0,50,50),
        'd' : np.random.randn(50)
}
data['b'] = data['a'] + 10*np.random.randn(50)
data['d'] = np.abs(data['d'])*100

plt.figure(figsize=(4,3))
plt.scatter('a','b',c='c',s='d',data=data) #plt.scatter(x,y,color,size,data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
names = ['group a','group b','group c']
values = [1,10,100]

plt.figure(figsize = (9,3))
plt.subplot(131) # (rowcolumnplotno)
plt.bar(names,values)
plt.subplot(132)
plt.scatter(names,values)
plt.subplot(133)
plt.plot(names,values)
plt.suptitle('Categorical Plotting')
plt.show()
```

```python
mu, sigma = 100 ,15
x = mu + sigma * np.random.randn(10000)

plt.figure(figsize=(5,3))

#histogram of data
n,bins,patches = plt.hist(x,50,density=True,facecolor='g',alpha=0.75)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60,0.025,r'$\mu=100,\ \sigma=15$')
plt.axis([40,160,0,0.03])
plt.grid(True)
plt.show()
```
